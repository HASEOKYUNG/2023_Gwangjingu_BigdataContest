### 1인 가구 시각화

```python
pop.query('년도==2020')[['자치구','one비율']].sort_values(by='one비율', ascending=False)\
.plot.bar(x='자치구', y='one비율', title="자치구별 1인가구 비율(2020년)",
          color=['lightgray']*3+['green']+['lightgray']*21, legend=False)
plt.xlabel('')
plt.show()
```

```python
pop.query('년도==2020')[['자치구','면적대비one']].sort_values(by='면적대비one', ascending=False)\
.plot.bar(x='자치구', y='면적대비one', title="자치구별 면적대비 1인가구수(2020년)",
          color=['lightgray']*3+['green']+['lightgray']*21, legend=False)
plt.xlabel('')
plt.show()
```

```python
fig, ax = plt.subplots(1,1)
pop.query('자치구=="광진구"')[['년도','가구수']].set_index('년도')\
.plot.bar(legend=False, title="광진구 1인가구수(2016~2020년)", color='lightgray', ax=ax)
pop.query('자치구=="광진구"')[['년도','one']].set_index('년도')\
.plot.bar(legend=False, color='green', ax=ax)
plt.xticks(rotation=0)
plt.xlabel('')
plt.show()
```

```python
pop.query('자치구=="광진구"')[['년도','one']].set_index('년도')\
.plot(title="광진구 1인가구수(2016~2020년)", legend=False, color='green', marker='o')
plt.xticks(range(2016,2021), rotation=0)
plt.xlabel('')
plt.show()
```
